# Books Read
A nice little list of books that I have read by Stephen King.

```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib
read_txt = 'king_books_read.txt'
read_list = [line.rstrip('\n') for line in open(read_txt)]
```

```python


wiki = "https://en.wikipedia.org/wiki/Stephen_King_bibliography"
soup = BeautifulSoup(urllib.request.urlopen(wiki), "lxml") #lxml gets rid of error
tables = soup.findAll("table", { "class" : "wikitable sortable" })
table = tables[0]

Books = dict() 
for row in table.findAll("tr"):
    cells = row.findAll("td")
    titles = row.findAll("i")
    if not cells:
        pass
    if cells:
        if len(cells) == 4:
            Books[titles[0].find(text=True)] = cells[2].find(text=True)
        if len(cells) == 5:
            Books[titles[0].find(text=True)] = cells[3].find(text=True)
        if len(cells) == 6:
            Books[titles[0].find(text=True)] = cells[4].find(text=True)

book_list = []
page_number_list = []
for key in Books:
    book_list.append(key)
    page_number_list.append(int(Books[key]))
```

```python
#Get index of books I have read
book_list_read = []
for book in read_list:
    if book in book_list:
        book_list_read.append(book_list.index(book))
```

```python
#Create read list from book_list
read_bin = []
for i in range(len(book_list)):
    if i in book_list_read:
        read_bin.append("Read")
    else:
        read_bin.append("")
```

```python
DataSet = list(zip(book_list,read_bin,page_number_list))
df = pd.DataFrame(data = DataSet, columns=['Book Title', 'Read or Not', "Number of Pages"])
df
```

```python
print('I have read '+str(len(book_list_read))+' books')
print("So I have "+str(len(book_list)-len(book_list_read))+" more books to read")
```

# When Will I Finish?

```python
number_of_pages_read = 0
for index, row in df.iterrows():
    if row['Read or Not'] == 'Read':
        number_of_pages_read += row["Number of Pages"]
total_pages = sum(page_number_list)
page_percentage = number_of_pages_read/total_pages
```

```python
from datetime import datetime
from dateutil.relativedelta import relativedelta

def diff_month(d1, d2):
    return (d1.year - d2.year) * 12 + d1.month - d2.month

day_I_started = datetime(2017,9,8)
months_since_started = diff_month(datetime.today(), day_I_started)

page_rate_per_month = number_of_pages_read/months_since_started

print("I am reading approximately "+str(int(page_rate_per_month))+" pages per month")

months_until_finished = int(total_pages/page_rate_per_month)

finishing_date = day_I_started+ relativedelta(months=months_until_finished)
print("I will finish with Stephen King books in "+str(finishing_date.strftime("%B"))+", "+str(finishing_date.strftime("%Y")))
```
